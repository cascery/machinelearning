```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score

# Load the dataset
train_data = np.load("TrainStruct3.npy")
test_data = np.load("TestStruct3.npy")

# Scale train and test separately to avoid data leakage
scaler = StandardScaler()
train_scaled = scaler.fit_transform(train_data)  # Fit only on train
test_scaled = scaler.transform(test_data)        # Transform test separately

# Combine train and test after scaling
data_scaled = np.vstack((train_scaled, test_scaled))

# Plot dendrogram to determine an initial range for clusters
plt.figure(figsize=(10, 5))
plt.title("Dendrogram")
linkage_matrix = sch.linkage(train_scaled, method='ward')  # Only use train data
sch.dendrogram(linkage_matrix)
plt.xlabel("Samples")
plt.ylabel("Distance")
plt.show()

# Automatically determine the best number of clusters using Silhouette Score
best_score = -1
best_clusters = 2  # Start with at least 2 clusters

for k in range(2, 11):  # Testing cluster numbers from 2 to 10
    cluster_model = AgglomerativeClustering(n_clusters=k, linkage='ward')
    labels_train = cluster_model.fit_predict(train_scaled)  # Cluster only on train
    score = silhouette_score(train_scaled, labels_train)
    print(f"Clusters: {k}, Silhouette Score: {score:.4f}")

    if score > best_score:
        best_score = score
        best_clusters = k

print(f"Optimal number of clusters: {best_clusters}")

# Perform final clustering on both train and test (but fit only on train)
cluster_model = AgglomerativeClustering(n_clusters=best_clusters, linkage='ward')
labels_train = cluster_model.fit_predict(train_scaled)  # Train clusters
labels_test = cluster_model.fit_predict(test_scaled)  # Test clusters

# Combine labels while keeping the original train-test split
labels = np.concatenate((labels_train, labels_test))

# Visualize clusters
plt.figure(figsize=(8, 6))
plt.scatter(data_scaled[:, 0], data_scaled[:, 1], c=labels, cmap='rainbow', edgecolors='k')
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.title("Agglomerative Clustering")
plt.show()

# Save the cluster labels
np.save("cluster_labels.npy", labels)

# Check if clustering is fair or just separating train/test
train_size = train_data.shape[0]
train_labels = labels[:train_size]
test_labels = labels[train_size:]

print("Train Cluster Distribution:", np.unique(train_labels, return_counts=True))
print("Test Cluster Distribution:", np.unique(test_labels, return_counts=True))
```

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score
from scipy.spatial.distance import cdist

#  Load the dataset
train_data = np.load("TrainStruct3.npy")
test_data = np.load("TestStruct3.npy")

#  Scale train and test separately to avoid data leakage
scaler = StandardScaler()
train_scaled = scaler.fit_transform(train_data)  # Fit only on train
test_scaled = scaler.transform(test_data)        # Transform test separately

#  Plot dendrogram for cluster analysis
plt.figure(figsize=(10, 5))
plt.title("Dendrogram (Train Data)")
linkage_matrix = sch.linkage(train_scaled, method='ward')
print(linkage_matrix)   # Only use train data
sch.dendrogram(linkage_matrix)
plt.xlabel("Samples")
plt.ylabel("Distance")
plt.show()

# 🔍 Determine the best number of clusters using Silhouette Score
best_score = -1
best_clusters = 2  # Minimum number of clusters

for k in range(2, 11):  # Testing cluster numbers from 2 to 10
    cluster_model = AgglomerativeClustering(n_clusters=k, linkage='ward')
    labels_train = cluster_model.fit_predict(train_scaled)  # Cluster only on train
    score = silhouette_score(train_scaled, labels_train)
    print(f"Clusters: {k}, Silhouette Score: {score:.4f}")

    if score > best_score:
        best_score = score
        best_clusters = k

print(f"✅ Optimal number of clusters: {best_clusters}")

#  Train clustering only on the training data
cluster_model = AgglomerativeClustering(n_clusters=best_clusters, linkage='ward')
labels_train = cluster_model.fit_predict(train_scaled)  # Train on train only

#  Find cluster centroids in training data
cluster_centroids = np.array([train_scaled[labels_train == i].mean(axis=0) for i in range(best_clusters)])

#  Assign test samples to the closest cluster using Euclidean distance
distances = cdist(test_scaled, cluster_centroids)
labels_test = np.argmin(distances, axis=1)  # Assign to the nearest centroid

labels = np.concatenate((labels_train, labels_test))

#  Save the cluster labels
np.save("cluster_labels.npy", labels)

#  Check cluster distribution
print("Train Cluster Distribution:", np.unique(labels_train, return_counts=True))
print("Test Cluster Distribution:", np.unique(labels_test, return_counts=True))

#  Visualize train vs test clustering
plt.figure(figsize=(8, 6))

# Plot train data
plt.scatter(train_scaled[:, 0], train_scaled[:, 1], c=labels_train, cmap='rainbow', edgecolors='k', label="Train Data")

# Plot test data
plt.scatter(test_scaled[:, 0], test_scaled[:, 1], c=labels_test, cmap='coolwarm', edgecolors='k', marker='x', label="Test Data")

plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.title("Hierarchical Clustering - Train vs Test")
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import os
from sklearn.cluster import KMeans
from PIL import Image

def load_data():
    A1 = np.load("A1.npy")
    A2 = np.load("A2.npy")
    return A1, A2

def perform_clustering(data, n_clusters=5):
    kmeans = KMeans(n_clusters=n_clusters, random_state=0, n_init=10)
    labels = kmeans.fit_predict(data)
    return labels, kmeans.cluster_centers_

def plot_clusters_with_images(data, labels, image_paths):
    plt.figure(figsize=(10, 10))
    unique_labels = np.unique(labels)
    colors = plt.cm.get_cmap("tab10", len(unique_labels))
    
    for idx, (point, label) in enumerate(zip(data, labels)):
        img_path = image_paths[idx]
        if os.path.exists(img_path):
            img = Image.open(img_path)
            img = img.resize((30, 30))
            plt.imshow(img, extent=(point[0] - 0.015, point[0] + 0.015, point[1] - 0.015, point[1] + 0.015), zorder=1)
        
        plt.scatter(point[0], point[1], color=colors(label), edgecolors='black', zorder=2)
    
    plt.title("Cluster Visualization with Images")
    plt.show()

def get_image_paths(A, database_path="database"):
    image_paths = []
    for row in A:
        person_id = int(row[-2])
        photo_id = int(row[-1])
        img_path = os.path.join(database_path, str(person_id), f"{photo_id}.bmp")
        image_paths.append(img_path)
    return image_paths

A1, A2 = load_data()
labels, centers = perform_clustering(A1[:, :-2])  # Clustering without last two columns (IDs)
image_paths = get_image_paths(A1)
plot_clusters_with_images(A1[:, :-2], labels, image_paths)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as sch

# 🚀 Load the dataset
train_data = np.load("TrainStruct3.npy")

# ✅ Scale data
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
train_scaled = scaler.fit_transform(train_data)

# 📊 Compute linkage matrix
linkage_matrix = sch.linkage(train_scaled, method='ward')

# 🔥 Find the biggest "jump" in linkage distances
distances = linkage_matrix[:, 2]  # Third column is the linkage distance
gaps = np.diff(distances)  # Compute differences between consecutive distances
best_cut = np.argmax(gaps)  # Find the largest gap

# ✅ The number of clusters = number of remaining branches **after** this biggest cut
num_clusters = len(distances) - best_cut
print(f"✅ Estimated Number of Clusters: {num_clusters}")

# 🎨 Plot the dendrogram
plt.figure(figsize=(10, 5))
plt.title("Dendrogram (Train Data)")
sch.dendrogram(linkage_matrix)
plt.xlabel("Samples")
plt.ylabel("Distance")
plt.show()
```
